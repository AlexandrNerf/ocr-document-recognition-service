```python
import sys
import os
import numpy as np
import plotly.graph_objects as go
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))

from shift_ocr.loaders.image_loader import ImageLoader
from shift_ocr.recognizer.easy_ocr import EasyOCRRecognizer
from shift_ocr.utils.read_detection import read_detections
from shift_ocr.postprocessor.simple import SimplePostprocessor
from shift_ocr.utils.BBoxVisualizer import BoundingBoxVisualizer
```

```python
def bgr_to_rgb(image: np.array) -> np.array:
    return image[..., ::-1]


image_path = '..\\assets\\images'
data_path = '..\\assets\\data'

loader = ImageLoader(image_path).load()
images = [np.array(bgr_to_rgb(image)) for image in loader]
txt_files = [os.path.join(data_path, f) for f in os.listdir(data_path)]

type_bbox = 'yolo' # либо 'pascal_voc'
```

```python
_, detection = read_detections(images, data_path, box_format=type_bbox)
detection1 = detection[0][0]
detection1
```

```python
BoundingBoxVisualizer.show_image(images[0], detection[0])
```

```python
recog_easy_ocr = EasyOCRRecognizer(
    lang_list = ['en', 'ru', 'tjk'], # тут 'kz' это кастомный словарь символов при его наличии качесвто чуть повышается(для запуска без ошибки его надо убрать)
    cuda = False,
    model_storage_directory=None,
    user_network_directory=None,
    verbose=False,
    quantize=False,
    download_enabled=True
)
```

```python
postprocessor = SimplePostprocessor()
post_detections = postprocessor.postprocessing(detection[0], images[0])
results = recog_easy_ocr.recognize(post_detections)

for res in results:
    print(res.text)
```
