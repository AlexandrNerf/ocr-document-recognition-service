```python
import matplotlib.pyplot as plt
import cv2
import sys
import torch

sys.path.append('../../')
from shift_ocr.detectors.fast import FASTDetector
from shift_ocr.utils.BBoxVisualizer import BoundingBoxVisualizer
from shift_ocr.postprocessor.simple import SimplePostprocessor
```

### С помощью детектора получаем предикты и смотрим на визуализацию

```python
detector = FASTDetector(
            cuda=True,
            pretrained=True,
            assume_straight_pages=False
        )
postprocessor = SimplePostprocessor()
img = cv2.imread('img.png')

print(img.shape)
plt.imshow(img)
```

```python
detections = detector.detect(img)
detections = postprocessor.postprocessing(detections, img)
print(len(detections))
```

```python
BoundingBoxVisualizer.show_image(img, detections)
```

Вырежем бокс

```python
for det in detections:
    BoundingBoxVisualizer.show_crop(img, det)
```
